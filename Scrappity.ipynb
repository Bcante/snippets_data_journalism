```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.stem.snowball import FrenchStemmer
from nltk import ngrams
from textblob import Word
from sklearn.feature_extraction.text import CountVectorizer

stemmer = FrenchStemmer()
stop = stopwords.words('french')
```

```python
df = pd.read_csv('archives_traites.csv', encoding='latin-1')
df.head(5)
```

```python

#lowercase, ponctuation, stop words
df['titre_stopwords'] = df['titre'].apply(lambda x: " ".join(x.lower() for x in x.split())) 
df['titre_stopwords'] = df['titre_stopwords'].str.replace('[^\w\s]',' ') 
df['titre_stopwords'] = df['titre_stopwords'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))

df['titre_stem'] = df['titre_stopwords'].apply(lambda x: " ".join([stemmer.stem(word) for word in x.split()]))
df.head(5)
```

```python
word_vectorizer = CountVectorizer(ngram_range=(2,3), analyzer='word')
sparse_matrix = word_vectorizer.fit_transform(df['titre_stem']) 
frequencies = sum(sparse_matrix).toarray()[0]
res = pd.DataFrame(frequencies, index=word_vectorizer.get_feature_names(), columns=['frequency'])
res.sort_values(by=['frequency'], ascending=False)
```
